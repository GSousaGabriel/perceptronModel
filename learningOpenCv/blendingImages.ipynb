```python
img1 = cv2.imread('../ACARMANBOSS.jpg')
img1= cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('../ABANDITLORD.jpg')
img2= cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2= cv2.resize(img2, (600, 600))
img1= cv2.resize(img1, (934, 1401))
```

```python
img1.shape
```

```python
x_offset = 934 - 600
y_offset = 1401 - 600
```

```python
rows, cols, channels= img2.shape
```

```python
roi= img1[y_offset:1401, x_offset:934]
```

```python
plt.imshow(roi)
```

```python
img2gray= cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img2gray, 'gray')
```

```python
mask_inv= cv2.bitwise_not(img2gray)
```

```python
plt.imshow(mask_inv, 'gray')
```

```python
mask_inv.shape
```

```python
white_bg= np.full(img2.shape, 255, np.uint8)
```

```python
white_bg.shape
```

```python
bk= cv2.bitwise_or(white_bg, white_bg, mask_inv)
```

```python
plt.imshow(bk)
```

```python
fg= cv2.bitwise_or(img2, img2, mask_inv)
```

```python
plt.imshow(fg)
```

```python
final_roi= cv2.bitwise_or(roi, fg)
```

```python
plt.imshow(final_roi)
```

```python
large_img= img1
small_img= final_roi
```

```python
large_img[y_offset:y_offset+small_img.shape[0], x_offset:x_offset+small_img.shape[1]]= small_img
```

```python
plt.imshow(large_img)
```
