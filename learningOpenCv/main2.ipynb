```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
blank_img= np.zeros((512,512, 3), np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, (0,100), (400, 200), (0,255,255), 10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, (200,200), (300, 300), (0,255,255), 10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, (100, 100), 50, (255,0,255), 10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, (400, 300), 50, (255,0,255), -1)
```

```python
plt.imshow(blank_img)
```

```python
cv2.line(blank_img, (0,0), (512,512), (255,255,0), 3)
```

```python
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, 'hello', (20, 490), font, 3, (124, 124, 543), 3, cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
blank_img2= np.zeros((512,512,3), np.int32)
plt.imshow(blank_img2)
```

```python
vertices = np.array([ [100, 300], [300,200], [400,300], [200, 400] ])
```

```python
pts= vertices.reshape((-1, 1, 2))
pts.shape
```

```python
cv2.polylines(blank_img2, [pts], True, (255, 300, 122), 3)
plt.imshow(blank_img2)
```
