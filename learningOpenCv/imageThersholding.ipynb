```python
img= cv2.imread('../ACARMANBOSS.jpg', 0)
plt.imshow(img)
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax= fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th1= cv2.threshold(img, 180, 255, cv2.THRESH_BINARY)
plt.imshow(th1)
```

```python
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 7, 6)
show_pic(th2)
```

```python
blended = cv2.addWeighted(th1, 0.6, th2, 0.4, 0)
show_pic(blended)
```
