```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img= cv2.imread('../ACARMANBOSS.jpg')
plt.imshow(img)
```

```python
fixedImg= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fixedImg)
```

```python
flippedImg= cv2.flip(fixedImg, 0)
plt.imshow(flippedImg)
```

```python
cv2.rectangle(fixedImg, (340, 400), (380, 320), (255,0,0), 2)
plt.imshow(fixedImg)
```

```python
fixedImg= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
points = np.array([ [340, 400], [380, 400], [360, 330], [340, 400]  ])
cv2.polylines(fixedImg, [points], False, (255, 0, 0), 2)
cv2.fillPoly(fixedImg, [points], 255)
plt.imshow(fixedImg)
```
